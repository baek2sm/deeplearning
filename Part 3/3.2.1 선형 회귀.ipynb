파이토치로 선형 회귀를 실습합니다.

```
# 실습에 필요한 라이브러리를 불러옵니다.
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
import torch
import torch.nn as nn
import torch.optim as optim
```

```
# 사이킷런에서 보스턴 주택 가격 데이터 세트를 불러옵니다.
dataset = datasets.load_boston()

# 입력 데이터와 타깃을 준비합니다.
X, y = dataset['data'], dataset['target']
```

```
# 입력 데이터와 타깃을 텐서 자료구조로 변환합니다.
X = torch.FloatTensor(X)
y = torch.FloatTensor(y).unsqueeze(-1)
```

```
# 데이터 범위 조정을 위해 표준화를 적용합니다.
X = (X - torch.mean(X)) / torch.std(X)
```

```
# 13개 값을 입력받고 1개의 값을 출력하는 선형 회귀 모델을 정의합니다.
model = nn.Linear(13, 1)
```

```
# 평균 제곱 오차(Mean Squared Error) 손실 함수 객체를 생성합니다.
criterion = nn.MSELoss()

# Adam 옵티마이저 객체를 생성합니다.
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

```
# 학습 함수를 정의합니다.
def train(model, criterion, optimizer, X, y):
  # 기울기를 초기화합니다.
  optimizer.zero_grad()
  # 모델을 사용해 결과를 예측합니다.
  hypothesis = model(X)
  # 오차를 계산합니다.
  loss = criterion(hypothesis, y)
  # 기울기를 계산합니다.
  loss.backward()
  # 경사 하강법으로 가중치를 수정합니다.
  optimizer.step()  
  # 현재 에포크의 오차를 반환합니다.
  return loss.item()
```

```
# 100회에 걸쳐 모델을 학습시킵니다.
n_epochs = 100
for epoch in range(1, n_epochs+1):
  # 모델을 학습시킵니다.
  loss = train(model, criterion, optimizer, X, y)  
  # 10 epoch마다 오차를 출력합니다.
  if epoch % 10 == 0:
    print('epoch: {}, loss: {:.4f}'.format(epoch, loss))
```
