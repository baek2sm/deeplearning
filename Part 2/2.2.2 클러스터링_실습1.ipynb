```python
# 실습에 필요한 패키지들을 불러옵니다.
import numpy as np
import matplotlib.pyplot as plt

# 3개의 클러스터로 분류할 임의의 데이터 세트를 만듭니다.
X = np.array([[1.2, 1.3], [1.1, 1.0], [0.8, 1.1],
              [4.5, 4.4], [4.0, 4.5], [4.8, 3.9],
              [9.9, 9.8], [8.9, 9.5], [9.2, 9.1]])

# 데이터 분포를 시각화하기 위해 산점도 그래프를 그립니다.
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
# 2행 1열 데이터 선택
print(X[1, 0])
```

```python
# 1행부터 5행까지 1번째 열만 선택
print(X[0:5, 0])

# 1행부터 5행까지 1번째 열만 선택
print(X[:5, 0])
```

```python
# 모든 행의 1번째 열만 선택
print(X[:, 0])

# 모든 행의 2번째 열만 선택
print(X[:, 1])
```
