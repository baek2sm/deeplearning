```python
# 실습에 필요한 라이브러리를 불러옵니다.
from sklearn import datasets
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
# 사이킷런에서 붓꽃 품종 데이터 세트를 불러옵니다.
dataset = datasets.load_iris()

# 입력 데이터와 타깃을 준비합니다.
X, y = dataset.data, dataset.target
```

```python
# 랜덤 포레스트 모델 객체를 생성합니다.
rf = RandomForestClassifier()

# 그리드 서치로 시도해 볼 하이퍼 파라미터 조합을 준비합니다.
param_grid = [{'n_estimators': range(5, 51, 5), 'max_features': range(1, 5), 'max_depth': range(1, 6)}]
```

```python
# 그리드 서치 객체를 생성합니다.
gs = GridSearchCV(estimator=rf, param_grid=param_grid, scoring='accuracy', cv=5, n_jobs=-1)

# 그리드 서치로 모델을 학습합니다.
gs.fit(X, y)
```

```python
# 그리드 서치의 학습 결과를 출력합니다.
print('최적의 하이퍼 파라미터: {0}'.format(gs.best_params_))
print('최적의 하이퍼 파라미터일 때 정확도: {0:.2f}'.format(gs.best_score_))

# 최적화 모델을 가져옵니다.
model = gs.best_estimator_
```
