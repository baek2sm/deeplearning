```python
# 실습에 필요한 라이브러리를 불러옵니다.
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 3개의 클러스터로 분류할 임의의 데이터 세트를 만듭니다.
X = np.array([[1.2, 1.3], [1.1, 1.0], [0.8, 1.1],
                  [4.5, 4.4], [4.0, 4.5], [4.8, 3.9],
                  [9.9, 9.8], [8.9, 9.5], [9.2, 9.1]])
```

```python
# K-평균 군집 알고리즘을 이용해 모델을 학습시킵니다.
model = KMeans(n_clusters=3, random_state=0)
model.fit(X)
```

```python
# 데이터 샘플들이 각각 어떤 클러스터로 분류되었는지 확인합니다.
print(model.labels_)
```

```python
# 데이터 분포를 시각화하기 위해 산점도 그래프를 그립니다.
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='brg')
plt.show()
```

```python
# 새로운 데이터 샘플을 추가합니다.
X_test = np.array([[4.4, 4.0], [8.9, 7.8]])

# 데이터 분포를 시각화하기 위해 산점도 그래프를 그립니다.
plt.scatter(X[:, 0], X[:, 1], c=model.labels_, cmap='brg')
plt.scatter(X_test[0, 0], X_test[0, 1], color='orange')
plt.show()
```

```python
# 새로운 데이터 샘플에 대한 예측을 실행하고 결과를 확인합니다.
result = model.predict(X_test)
print('새로운 데이터 샘플이 속하는 클러스터 예측: {}'.format(result))
```

```python
# 학습된 K-평균 클러스터링 모델의 이너셔값을 확인합니다.
print('{:.2f}'.format(model.inertia_))
```
