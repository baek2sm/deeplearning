```python
# 고양이와 개 데이터 세트를 다운받고 압축을 해제합니다.
!wget https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip
!unzip cats_and_dogs_filtered.zip

# 맷플롯립 라이브러리와 OpenCV 라이브러리를 불러옵니다.
import matplotlib.pyplot as plt
import cv2
```

```python
# 흑백 이미지를 불러옵니다.
img_path = './cats_and_dogs_filtered/validation/dogs/dog.2000.jpg'
img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

# 이미지의 크기를 변경합니다.
img_resized = cv2.resize(img, (100, 100))

# 변경된 이미지를 저장합니다.
cv2.imwrite("./img_resized.jpg", img_resized)

# 저장한 이미지를 다시 불러옵니다.
img_loaded = cv2.imread("./img_resized.jpg", cv2.IMREAD_GRAYSCALE)

# 불러온 이미지 크기를 출력합니다.
print(img_loaded.shape)

# 맷플롯립 라이브러리를 이용해 이미지를 출력합니다.
plt.imshow(img_loaded, cmap='gray')
plt.show()
```
