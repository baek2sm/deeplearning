```python
# 고양이와 개 데이터 세트를 다운받고 압축을 해제합니다.
!wget https://storage.googleapis.com/mledu-datasets/cats_and_dogs_filtered.zip
!unzip cats_and_dogs_filtered.zip

# 맷플롯립 라이브러리와 OpenCV 라이브러리를 불러옵니다.
import matplotlib.pyplot as plt
import cv2
```

```python
# 컬러 이미지를 BGR 형태로 불러옵니다.
img_path = './cats_and_dogs_filtered/validation/dogs/dog.2000.jpg'
img_color = cv2.imread(img_path, cv2.IMREAD_COLOR)
```

```python
# BGR 컬러 이미지를 RGB 컬러 이미지로 변환합니다.
img_color = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)
```

```python
# 불러온 이미지의 데이터 타입과 형태를 출력합니다.
print(type(img_color), img_color.shape)

# 맷플롯립 라이브러리를 이용해 불러온 이미지를 출력합니다.
plt.imshow(img_color)
plt.show()
```
