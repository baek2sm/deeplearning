```python
# 맷플롯립을 plt라는 별칭으로 불러옵니다.
import matplotlib.pyplot as plt
```

```python
# 차트를 그릴 데이터를 준비합니다.
data = [11, 14, 12, 9, 15, 17]

# 맷플롯립을 이용해 꺾은선 그래프를 정의합니다.
plt.plot(data)

# 그래프를 화면에 그립니다.
plt.show()
```

```python
# 차트를 그릴 x축의 데이터와 y축의 데이터를 준비합니다.
x_data = [7, 8, 9, 10, 11, 12]
y_data = [11, 14, 12, 9, 15, 17]

# 맷플롯립을 이용해 꺾은선 그래프를 정의합니다.
plt.plot(x_data, y_data)

# 그래프를 화면에 그립니다.
plt.show()
```

```python
# 차트를 그릴 x축의 데이터를 준비합니다.
x_data = [7, 8, 9, 10, 11, 12]

# 차트를 그릴 y축의 데이터 2개를 준비합니다.
y1_data = [11, 14, 12, 9, 15, 17]
y2_data = [6, 8, 15, 10, 12, 11]

# 맷플롯립을 이용해 꺾은선 그래프 2개를 정의합니다.
plt.plot(x_data, y1_data)
plt.plot(x_data, y2_data)

# 그래프를 화면에 그립니다.
plt.show()
```

```python
# 차트를 그릴 x축의 데이터를 준비합니다.
x_data = [7, 8, 9, 10, 11, 12]

# 차트를 그릴 y축의 데이터 2개를 준비합니다.
y1_data = [11, 14, 12, 9, 15, 17]
y2_data = [6, 8, 15, 10, 12, 11]

# 그래프의 제목을 정의합니다.
plt.title('Monthly sales')

# 맷플롯립을 이용해 꺾은선 그래프 2개를 정의합니다.
plt.plot(x_data, y1_data)
plt.plot(x_data, y2_data)

# 축 이름을 지정합니다.
plt.xlabel('Month')
plt.ylabel('Sales')

# 그래프를 화면에 그립니다.
plt.show()
```

```python
# 차트를 그릴 x축의 데이터를 준비합니다.
x_data = [7, 8, 9, 10, 11, 12]

# 차트를 그릴 y축의 데이터 2개를 준비합니다.
y1_data = [11, 14, 12, 9, 15, 17]
y2_data = [6, 8, 15, 10, 12, 11]

# 그래프의 제목을 추가합니다.
plt.title('Monthly sales')

# 화면에 그릴 꺾은선 그래프 2개를 추가합니다.
plt.plot(x_data, y1_data, label='laptop')
plt.plot(x_data, y2_data, label='desktop')

# 축 이름을 지정합니다.
plt.xlabel('Month')
plt.ylabel('Sales')

# 그래프 화면에 범례를 표시합니다.
plt.legend()

# 그래프를 화면에 그립니다.
plt.show()
```

```python
# 차트를 그릴 때 사용할 공부 시간과 성적 데이터를 준비합니다.
hours_studied = [5, 8, 10, 15, 20]
grades = [70, 80, 85, 90, 95]

# 산점도 그래프를 추가합니다.
plt.scatter(hours_studied, grades)

# 축 이름을 지정합니다.
plt.xlabel('Hours Studied')
plt.ylabel('Grades')

# 그래프를 화면에 그립니다.
plt.show()
```
