```python
# 사이킷런의 데이터를 사용하기 위해 데이터 세트 모듈을 불러옵니다.
from sklearn import datasets

# 사이킷런에서 붓꽃 품종 데이터 세트를 가져옵니다.
dataset = datasets.load_iris()
```

```python
# 이 데이터 세트가 가지고 있는 특성들의 이름을 출력합니다.
print('특성 이름:\n{}'.format(dataset['feature_names']))
```

```python
# 입력 데이터의 5개 행을 출력합니다.
print('입력 데이터:\n{}'.format(dataset['data'][:5]))
```

```python
# 타깃의 이름을 출력합니다.
print('타깃 이름:\n{}'.format(dataset['target_names']))
```

```python
# 타깃을 출력합니다.
print('타깃:\n{}'.format(dataset['target']))
```
